# 02 - Sentinel Value Parsing & Preprocessing

Financial datasets use values like `-1` or `-999` to indicate "Not Applicable" for specific products (e.g., LTV for a Credit Card). If fed directly into an ML model, these will skew numerical calculations. We isolate them here.

```python

import pandas as pd
import numpy as np
import yaml

df = pd.read_csv('../data/raw/banking_finance_01.csv')
with open('../config/business_rules.yaml', 'r') as file:
    rules = yaml.safe_load(file)

print("Rules loaded successfully.")



print("Rules loaded successfully.")
```

### Handling -1 in LTV for Unsecured Products

```python
ltv_sentinel = rules['sentinel_values']['loan_to_value_pct']

# Create a flag indicating whether the field is applicable
df['ltv_is_applicable'] = np.where(df['loan_to_value_pct'] == ltv_sentinel, 0, 1)

# Replace the sentinel with NaN for numerical calculations
df['loan_to_value_pct_clean'] = df['loan_to_value_pct'].replace(ltv_sentinel, np.nan)

print("Cleaned LTV Summary:")
print(df['loan_to_value_pct_clean'].describe())
```

### Handling -999 in Property Value Changes for Non-Mortgages

```python
prop_sentinel = rules['sentinel_values']['property_value_change_pct_last_qtr']
df['property_val_applicable'] = np.where(df['property_value_change_pct_last_qtr'] == prop_sentinel, 0, 1)
df['property_val_change_clean'] = df['property_value_change_pct_last_qtr'].replace(prop_sentinel, np.nan)

print("\nCleaned Property Value Change Summary:")
print(df['property_val_change_clean'].describe())
```
