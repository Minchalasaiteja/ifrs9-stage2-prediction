# 04 - Model Training & Comparison

Training the XGBoost and LightGBM models to predict `stage_2_migration` based on engineered features.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import xgboost as xgb

# Placeholder: In reality load from data/processed/ifrs9_features_final.csv
df = pd.read_csv('../data/raw/banking_finance_01.csv') 

# Minimal viable feature set for demonstration
features = ['pd_relative_change_pct', 'days_past_due_current', 'internal_credit_score', 
            'credit_score_change_last_quarter', 'missed_payments_last_12m']
target = 'stage_2_migration'

X = df[features].fillna(0) # Simple imputation for demo
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### Train XGBoost

```python
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', scale_pos_weight=3.5)
xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)
y_prob = xgb_model.predict_proba(X_test)[:, 1]

print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
